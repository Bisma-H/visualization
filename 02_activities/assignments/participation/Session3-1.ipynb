```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1)
ax.plot(x,y2)
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label =
"Person 1" )
ax.plot(x,y2, label =
"Person 2" )
ax.legend(loc=
'lower right')
fig.show()
```
