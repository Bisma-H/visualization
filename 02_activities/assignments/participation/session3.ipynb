```python
#loading libraries

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy
import PIL
import requests
```

```python
#Generating random numbers for the y values to plot against the x values. The x values range from 0 to 50.
# Setting seed to make it reproducible

np.random.seed(613)
x=np.arange(50)
y = np.random.randint(0, 100, 50)


# Creating a scatter plot

fig, ax= plt.subplots(figsize= (5,3))
ax.scatter(x,y)
```

```python
#Creating a bar chart using the same data

fig, ax= plt.subplots(figsize= (5,3))
ax.bar(x,y)
```

```python

#Making th bar char horizontal

fig, ax= plt.subplots(figsize= (5,3))
ax.barh(x,y)
```

```python
# Creating a line plot using the same data

fig, ax= plt.subplots(figsize= (5,3))
ax.plot(x,y)
```

```python
#Creating a histogram using th y values generated

fig, ax= plt.subplots(figsize= (5,3))
ax.hist(y)
```

```python
# Adding titles to the graph and axis

fig, ax= plt.subplots(figsize= (5,3))
ax.plot(x,y)

ax.set_title('Total Growth Over Time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout
```

```python
# Creating  font dictionaries

font1 = {'family': 'sans-serif','color':'blue','size': 20}
font2 = {'family': 'monospace','color':'green','size': 14}
```

```python
fig, ax= plt.subplots(figsize= (5,3))
ax.plot(x,y)

ax.set_title('Total Growth Over Time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout
```

```python
fig, ax= plt.subplots(figsize= (5,3))
ax.plot(x,y)

ax.set_title('Total Growth Over Time', fontdict = font1, loc='left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout
```

```python
#Adding markers

fig, ax= plt.subplots(figsize= (5,3))
ax.scatter(x,y, marker='*', color = 'indigo')
```

```python
#Adding lines tot he graph

fig, ax= plt.subplots(figsize= (5,3))
ax.plot(x,y, marker='*', color = 'indigo', linestyle = '--', linewidth = 2)
fig.show()
```

```python
# Changing the color of the markers and the lines 

fig, ax= plt.subplots(figsize= (5,3))
ax.plot(x,y, marker='*', 
    markersize = 12,
    color = '#7425b9', 
    linestyle = '--', 
    linewidth = 2,
    markeredgecolor = '#fa9359',
    markerfacecolor = '#fa9359')
fig.show()
```

```python
fig, ax= plt.subplots(figsize= (5,3))
ax.plot(x,y, marker='*', 
    markersize = 12,
    color = '#7425b9', 
    linestyle = '--', 
    linewidth = 2,
    markeredgecolor = '#fa9359',
    markerfacecolor = '#000000')
ax.grid()
```

```python


# import libraries
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# load dataset
tips = sns.load_dataset("tips")

# Set the figure size
plt.figure(figsize=(14, 8))

# plot a bar chart
ax = sns.barplot(x="day", y="total_bill", data=tips, estimator=np.mean, ci=85, capsize=.2, color='lightblue')
```

This graph shows the different bill amounts for the days Thursday - Sunday. We can see that the bill is slighlty higher on the weekend. Aesthetically, the graph could look better but for an academic audience it would be appropriate. There should be a title telling us what exactly 'bill' here means and the currency it is being measured in. In terms of substance, it conveys the idea that the weekends expenditure is higher than the weekdays. The error bars also tell us how precise these estimates are.
